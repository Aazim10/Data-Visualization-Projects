```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_imdb_1000 = pd.read_csv('imdb_top_1000.csv')
df_imdb_1000.head()
```

```python
df_imdb_1000.info()
```

```python
df = df_imdb_1000[['Series_Title','Released_Year','Certificate','Runtime','Genre','IMDB_Rating','Meta_score','Director','Star1','Star2','Star3','Star4','No_of_Votes','Gross']]
df.info()
```

```python
df[df['Released_Year'].str.contains(r'[a-zA-Z]')]
```

```python
df.loc[966,'Released_Year'] = 1995
df['Released_Year'] = df['Released_Year'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df[df['Runtime'].str.replace('min','').str.strip().str.contains(r'[a-zA-Z]')]
```

```python
df['Runtime'] = df['Runtime'].str.replace('min','').str.strip()
df['Runtime'] = df['Runtime'].astype(int)
```

```python
df.info()
```

```python
df['Gross'] = df['Gross'].str.replace(',','')
df['Gross'] = df['Gross'].fillna(0)
df['Gross'] = df['Gross'].astype(int)
```

```python
df.describe()
```

```python
df.info()
```

```python
def genre_break(item):
    items = item.split(',')
    ss_items = []
    for i in items:
        ss_items.append(i.strip())
    return ss_items
        


df['Genre'] = df['Genre'].apply(genre_break)
```

```python
tot_gen_count=0
for row in df['Genre']:
    global tot_gen_count
    tot_gen_count = tot_gen_count + len(row)

tot_gen_count
```

```python
genres = {'Drama'}
for row in df['Genre']:
    for gen in row:
        global genres
        genres.add(gen)
genres
```

```python
len(genres)
```

```python
df_genre_analysis = pd.DataFrame(columns=['Series_Title','Released_Year','Runtime','Genre','IMDB_Rating','Meta_score','Director','No_of_Votes','Gross'])
```

```python
for row in df.itertuples():
    for genre in row.Genre:
        df_genre_analysis = df_genre_analysis.append({'Series_Title':row.Series_Title,'Released_Year':row.Released_Year,'Runtime':row.Runtime,'Genre':genre,'IMDB_Rating':row.IMDB_Rating,'Meta_score':row.Meta_score,'Director':row.Director,'No_of_Votes':row.No_of_Votes,'Gross':row.Gross},ignore_index=True)

df_genre_analysis.head()
```

```python
df_genre_analysis['Released_Year'] = df_genre_analysis['Released_Year'].astype(int)
df_genre_analysis['Genre'] = df_genre_analysis['Genre'].astype('category')
df_genre_analysis['No_of_Votes'] = df_genre_analysis['No_of_Votes'].astype(int)
df_genre_analysis['Gross'] = df_genre_analysis['Gross'].astype(int)
df_genre_analysis['Runtime'] = df_genre_analysis['Runtime'].astype(int)
```

```python
df_genre_analysis.info()
```

## Genre Distribution: 
Create a bar chart that shows the distribution of movies across different genres. This will help you identify the most common genres.

```python
df_bar = df_genre_analysis['Genre'].value_counts().reset_index()
df_bar.head()
```

```python
fig, ax = plt.subplots(1,2,figsize=(10, 4),sharey=True)
bar1 = sns.barplot(x='Genre',y='index',data=df_bar,order=df_bar['index'],ax=ax[0])
bar2 = sns.barplot(x='Genre',y='index',data=df_bar,order=df_bar['index'],ax=ax[1])
bar3 = sns.catplot(x='Genre',y='index',data=df_bar,kind='bar',order=df_bar['index'])
bar1.set(xlabel ="No of Movies", ylabel = "Genres", title ='Most Popular Genres')
bar2.set_xlabel("No of Movies")
bar2.set_ylabel("Genres")
bar2.set_title("Most Popular Genres")
bar3.set(xlabel ="No of Movies", ylabel = "Genres", title ='Most Popular Genres')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(df_bar['index'],df_bar['Genre'])
ax.set_xticklabels(df_bar['index'],rotation=90)
ax.set_title('Most Popular Genres')
bar2.set_ylabel("No of Movies")
bar2.set_xlabel("Genres")
plt.show()
```

## Genre Popularity Over Time: 
Plot a line chart or area chart to visualize how the popularity of different genres has changed over the years. This can help you identify trends and shifts in audience preferences.

```python
df_line = df_genre_analysis.groupby(['Released_Year','Genre']).agg({'Series_Title':'count'})
df_line.head(10)
```

```python
line = sns.relplot(x='Released_Year',y='Series_Title',col='Genre',col_wrap=3,col_order=df_bar['index'],data=df_line,kind='line')
# line.set(title='Genre Popularity Over Time')
plt.show()
```

```python
df_line_pop = df_genre_analysis.groupby(['Released_Year','Genre']).agg({'No_of_Votes':'sum'})
line = sns.relplot(x='Released_Year',y='No_of_Votes',col='Genre',col_wrap=3,col_order=df_bar['index'],data=df_line_pop,kind='line')
# line.set(title='Genre Popularity Over Time')
plt.show()
```

```python
fig, ax = plt.subplots(7,3,figsize=(18,45),sharey=True)
df_line_mat = df_line_pop.reset_index()
genres = list(df_bar['index'])
g=0
for i in range(7):
    for j in range(3):
        df_gen = df_line_mat[df_line_mat['Genre']==genres[g]]
        ax[i,j].plot(df_gen['Released_Year'],df_gen['No_of_Votes'])
        ax[i,j].set(xlabel =genres[g],ylabel="No_of_Votes")
        g=g+1
    
plt.show()
```

## Average Rating by Genre: 
Create a bar chart or box plot to display the average ratings for each genre. This will give insights into which genres tend to have higher or lower ratings.

```python
df_rate_bar = df_genre_analysis.groupby('Genre').agg({'IMDB_Rating':'mean'}).reset_index().sort_values('IMDB_Rating')
df_rate_bar.head()
```

```python
rate_bar = sns.barplot(x='Genre',y='IMDB_Rating',data=df_genre_analysis,order=df_rate_bar['Genre'])
rate_bar.set(title ='Average Rating by Genre')
plt.xticks(rotation=90)
```

```python
fig, ax = plt.subplots()
ax.bar(df_rate_bar['Genre'],df_rate_bar['IMDB_Rating'],color='orange')
ax.set_xticklabels(df_rate_bar['Genre'],rotation=90)
ax.set_title('Average Rating by Genre')
ax.set_ylabel("IMDB_Rating")
ax.set_xlabel("Genres")
plt.show()
```

## Correlation Heatmap: 
Generate a correlation heatmap using Seaborn to visualize the correlations between attributes like ratings and release years. This can help you identify relationships between variables.

```python
df_heatmap = df[['Released_Year','Runtime','IMDB_Rating','Meta_score','No_of_Votes','Gross']]
df_heatmap.head()
```

```python
heat_map = sns.heatmap(df_heatmap.corr(),annot=True)
heat_map.set(title ='Correlation Heatmap')
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.imshow(df_heatmap.corr(), interpolation='nearest')
fig.colorbar(im, orientation='vertical', fraction = 0.05)
cols= ['']
cols.extend(df_heatmap.columns)
ax.set_xticklabels(cols, rotation=90,fontsize=10)
ax.set_yticklabels(cols, rotation=0, fontsize=10)
for i in range(len(df_heatmap.columns)):
    for j in range(len(df_heatmap.columns)):
        text = ax.text(j, i, round(df_heatmap.corr().to_numpy()[i, j],2),
                       ha="center", va="center", color="black")
plt.show()
```

## Top Movies by Genre: 
Create a horizontal bar chart showcasing the top-rated movies for each genre. This will highlight the most acclaimed movies within each genre.

```python
df_hor_bar = df_genre_analysis.groupby('Genre').agg({'IMDB_Rating':'max'}).reset_index()
df_hor_bar_new = df_hor_bar.merge(df_genre_analysis,on=['Genre','IMDB_Rating']).drop_duplicates('Genre').reset_index().sort_values('IMDB_Rating')
df_hor_bar_new['label'] = df_hor_bar_new['Genre'].astype(str)+' - '+df_hor_bar_new['Series_Title']
df_hor_bar_new.head()
```

```python
hor_bar = sns.barplot(x='Genre',y='IMDB_Rating',data=df_hor_bar_new,order=df_hor_bar_new['Genre'])
hor_bar.set_xticklabels(df_hor_bar_new['label'],rotation=90)
hor_bar.set(xlabel ="Genre - Movie", title ='Top Movies by Genre')
```

```python
fig, ax = plt.subplots()
ax.bar(df_hor_bar_new['Genre'],df_hor_bar_new['IMDB_Rating'],color='green')
ax.set_xticklabels(df_hor_bar_new['label'],rotation=90)
ax.set(xlabel ="Genre - Movie",ylabel="IMDB_Rating", title ='Top Movies by Genre')
plt.show()
```
